```python
from dataclasses import dataclass, field
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# Signal Data Structure

@dataclass
class Signal:
    # Waveform
    time: np.ndarray
    voltage: np.ndarray

    # Running metadata
    cumulative_loss_db: float = 0.0
    cumulative_delay_ns: float = 0.0
    component_log: list = field(default_factory=list)


# ----------------------------------------------------
# Sample PMT Waveform
t = np.linspace(0, 100, 1000)

# Gaussian pulse
v = np.exp(-((t - 30) / 5) ** 2)

# Signal object
signal = Signal(time=t, voltage=v)

# Record the source
signal.component_log.append("PMT Source")

# ----------------------------------------------------
# Signal Information
print(f"Cumulative Loss: {signal.cumulative_loss_db:.2f} dB")
print(f"Cumulative Delay: {signal.cumulative_delay_ns:.2f} ns")
print("Components Passed Through:")
for component in signal.component_log:
    print(f" - {component}")

# ----------------------------------------------------
# Plot
plt.figure(figsize=(8, 4))
plt.plot(signal.time, signal.voltage, label="PMT Waveform")
plt.title("Synthetic PMT Waveform")
plt.xlabel("Time (ns)")
plt.ylabel("Voltage (V)")
plt.grid(True)
plt.legend()
plt.show()
```
